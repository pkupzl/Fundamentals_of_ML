$\text{1. In Bayesian linear regression, we assume the following model:}$
$$y_{i} = x_{i}^{T}\beta + \epsilon_{i}$$
$\text{where}$ $\epsilon_{i} \sim N(0, \sigma^{2})$. $\text{Assuming we set the following prior distribution:}$
$$ p(\beta, \sigma^{2}) \propto \sigma^{-2}$$
$\text{Let X represent the design matrix, }X=\begin{bmatrix}
x_{1}^{T} \\
\vdots \\ 
x_{n}^{T}
\end{bmatrix}$,  $y = (y_{1},...,y_{n})^{T} 
\text{prove that:}$<br>
$\text{(1.1)}\quad  p(\beta| \sigma^{2},y) \sim N(\hat{\beta},V_{\beta}\sigma^{2})$, where $\hat{\beta}=(X^{T}X)^{-1}X^{T}y$, $V_{\beta}=(X^{T}X)^{-1}$<br>
$\text{(1.2)}\quad p(\sigma^{2}|y) \sim \text{Scaled Inv-}\chi^{2} (n-k,s^{2})$, where $s^{2}=\frac{1}{n-k}\sum_{i=1}^{n}(y_{i}-x_{i}^{T}\hat{\beta})^{2}$, $k$ is the number of parameters in $\beta$.

$\text{Question 1.1}$ <br>
$\text{Proof:}$
$$ 
   \begin{aligned}
   &p(\beta|\sigma^{2},y) = \frac{p(y|\sigma^{2},\beta)p(\beta|\sigma^{2})}{p(y|\sigma^{2})}\\
   &          =\frac{p(y|\sigma^{2},\beta)p(\beta,\sigma^{2})}{p(y,\sigma^{2})}\\
   &          \propto p(y|\sigma^{2},\beta)p(\beta,\sigma^{2})\\
   &          \propto exp\left (-\frac{1}{2\sigma^{2}}(y-X\beta)^{T}(y-X\beta)\right )\sigma^{-2}\\
   &          \propto exp\left(-\frac{1}{2\sigma^{2}}(y^{T}y-2y^{T}X\beta+\beta^{T}X^{T}X\beta)\right )\sigma^{-2}\\
   &          \propto exp\left(-\frac{1}{2\sigma^{2}}(\beta^{T}X^{T}X\beta-2\beta X^{T}y)\right )\\
   &          \propto exp\left(-\frac{1}{2\sigma^{2}}(\beta^{T}X^{T}X\beta-2\beta X^{T}X\hat{\beta})\right ),\hat{\beta}\equiv (X^{T}X)^{-1}X^{T}y\\
   &          \propto exp\left(-\frac{1}{2\sigma^{2}}(\beta^{T}X^{T}X\beta-2\beta X^{T}X\hat{\beta}+\hat{\beta}X^{T}X\hat{\beta})\right )\\
   &          \propto exp(-\frac{1}{2\sigma^{2}}(\beta-\hat{\beta})^{T}X^{T}X(\beta-\hat{\beta}))\\
   &          \propto exp(-\frac{1}{2}(\beta-\hat{\beta})^{T}(V_{\beta}\sigma^{2})^{-1}(\beta-\hat{\beta})) ,V_{\beta}\equiv (X^{T}X)^{-1}\\
   \end{aligned}
$$
$$
   \therefore p(\beta|\sigma^{2},y) \sim N(\hat{\beta},V_{\beta}\sigma^{2})
$$

$\text{Question 1.2}$<br>
$\text{From Question 1.1 we actually know that}$
$$
\begin{aligned}
   &(y-X\beta)^{T}(y-X\beta) \\
   & = [y-X\hat{\beta}+X(\hat{\beta}-\beta)]^{T}[y-X\hat{\beta}+X(\hat{\beta}-\beta)]\\
   & = (y-X\hat{\beta})^{T}(y-X\hat{\beta})+(\hat{\beta}-\beta)^{T}X^{T}X(\hat{\beta}-\beta)\\
\end{aligned}
$$
$$
\begin{aligned}
   & \therefore p(y|\beta, \sigma^{2})\\
   & = \frac{1}{(2\pi)^{n/2}\sigma^{n}}exp\left(-\frac{1}{2\sigma^{2}}(y-X\beta)^{T}(y-X\beta)\right)\\
   & \propto \frac{1}{\sigma^{n}}exp\left(-\frac{1}{2\sigma^{2}}(y-X\hat{\beta})^{T}(y-X\hat{\beta})\right)exp\left(-\frac{1}{2\sigma^{2}}(\hat{\beta}-\beta)^{T}X^{T}X(\hat{\beta}-\beta)\right)\\
   & = (\sigma^{2})^{-\frac{v}{2}}exp\left(-\frac{vs^{2}}{2\sigma^{2}}\right)(\sigma^{2})^{-\frac{n-v}{2}} exp\left(-\frac{1}{2\sigma^{2}}(\hat{\beta}-\beta)^{T}X^{T}X(\hat{\beta}-\beta)\right)\\
\end{aligned}
$$
$\text{where } v \equiv n-k, s^{2} \equiv \frac{1}{n-k}(y-X\hat{\beta})^{T}(y-X\hat{\beta})$
$$
\begin{aligned}
   & \therefore p(\sigma^{2}|y)\propto p(y,\sigma^{2})\\
   & = \int p(y|\beta,\sigma^{2})p(\beta,\sigma^{2})d\beta\\
   & \propto (\sigma^{2})^{-\frac{v}{2}-1}exp\left(-\frac{vs^{2}}{2\sigma^{2}}\right)(\sigma^{2})^{-\frac{n-v}{2}} \int_{\beta} exp\left(-\frac{1}{2\sigma^{2}}(\hat{\beta}-\beta)^{T}X^{T}X(\hat{\beta}-\beta)\right)d\beta\\
\end{aligned}
$$
$\text{Obviously}$
$$
\begin{aligned}
   &1 = \int_{\beta} p(\beta|\sigma^{2},y)d\beta
   &= (2\pi)^{-\frac{k}{2}}|V_{\beta}\sigma^{2}|^{-\frac{1}{2}}\int_{\beta} exp\left(-\frac{1}{2\sigma^{2}}(\hat{\beta}-\beta)^{T}X^{T}X(\hat{\beta}-\beta)\right)d\beta\\
\end{aligned}
$$
$$

\begin{aligned}
&\because V_{\beta} \in \mathbb{R}^{k\times k} \quad \therefore  |V_{\beta}\sigma^{2}|^{-\frac{1}{2}} = (\sigma^{2})^{-\frac{k}{2}}|V_{\beta}|^{-\frac{1}{2}}\\
&\therefore  \int_{\beta} exp\left(-\frac{1}{2\sigma^{2}}(\hat{\beta}-\beta)^{T}X^{T}X(\hat{\beta}-\beta)\right)d\beta \propto (\sigma)^{\frac{k}{2}}\\
&\therefore p(\sigma^{2}|y) \propto (\sigma^{2})^{-\frac{v}{2}-1}exp\left(-\frac{vs^{2}}{2\sigma^{2}}\right)(\sigma^{2})^{-\frac{n-v}{2}} (\sigma^{2})^{\frac{k}{2}}\\
& =(\sigma^{2})^{-\frac{v}{2}-1}exp\left(-\frac{vs^{2}}{2\sigma^{2}}\right)\\
\end{aligned}
$$
$$
\therefore  p(\sigma^{2}|y) \sim \text{Scaled Inv-}\chi^{2} (n-k,s^{2})
$$

$\text{2. Suppose we have a posterior distribution f(x) as follows:}$
$$
f(x) \propto e^{-\frac{x^{2}}{2}}(sin^{2}(6+x)+3cos^{2}(x)\cdot sin^{2}(4x)+1)
$$
$\text{(2.1) Please use importance sampling to sample from it and estimate E(X\^ 2) based on the samples.}$<br>
$\text{(2.2) Please use rejection sampling to sample from it, estimate E(X\^ 2) as well, and create a histogram of the samples.}$

```python
from sympy import *
import numpy as np
import math
from typing import Generator
# Create Linear Congruential Generator
def lcg(modulus: int, a: int, c: int, seed: int) -> Generator[int, None, None]:
    """Linear congruential generator."""
    while True:
        seed = (a * seed + c) % modulus
        yield seed
def generate_standard_normal_distributed_numbers(modulus: int, a: int, c: int, seed: int, num_samples: int):
    """Generate a list of numbers following standard normal distribution using LCG."""
    lcg_generator = lcg(modulus, a, c, seed)
    uniform_samples = [next(lcg_generator) / modulus for _ in range(num_samples)]
    
    # Applying Box-Muller transform to convert uniform samples to standard normal distribution
    normal_samples = []
    for i in range(0, num_samples, 2):
        u1 = uniform_samples[i]
        u2 = uniform_samples[i+1]
        z1 = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
        z2 = math.sqrt(-2 * math.log(u1)) * math.sin(2 * math.pi * u2)
        normal_samples.extend([z1, z2])
    
    return normal_samples
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
modulus = 2**31 - 1
a = 1103515245
c = 12345
seed = 0
num_samples = 1000000
# sample from standard normal distribution
random_numbers = generate_standard_normal_distributed_numbers(modulus, a, c, seed, num_samples)

# plot histogram
plt.hist(random_numbers, bins=120, density=True, alpha=0.7, color='g',label='Histogram')
# plot fitted curve
mu, std = norm.fit(random_numbers)
x = np.linspace(min(random_numbers), max(random_numbers), 100)
y = norm.pdf(x, mu, std)
plt.plot(x, y, 'r', label='Fitting Curve')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Fitted Normal Distribution')
plt.grid(True)
plt.legend()
plt.show()
print(f'mu = {mu}, std = {std}')
```

```python
lcg_generator = lcg(modulus, a, c, seed)
from scipy.special import erfinv
def q_sample():
    u = next(lcg_generator) / modulus
    return math.sqrt(2) * erfinv(2 * u - 1)
def q(x):
    """Calculate the value of q(x),where q is the standard Gauss distribution."""
    return 1 / math.sqrt(2 * math.pi) * np.exp(-x**2 / 2)
def f(x):
    """Calculate the value of f(x) = e^(-x^2/2)(sin^2(6+x) + 3cos^2(x) * sin^2(4x) + 1)"""
    return np.exp(-x**2/2) * (np.sin(6+x)**2 + 3*np.cos(x)**2 * np.sin(4*x)**2 + 1)
```

```python
# Determine M 
from scipy import optimize
def ratio(x):
    """Calculate the value of f(x)/q(x)"""
    return f(x) / q(x)

# Find the maximum of f(x)/q(x)
result = optimize.fminbound(lambda x: -ratio(x), -10, 10)
max_ratio = ratio(result)

print(f"The maximum value of f(x)/q(x) is {max_ratio} at x = {result}")
```

```python
# plot M*q(x) and f(x)
M = max_ratio*2
x = np.linspace(-5, 5, 1000)
Mq = M * q(x)
fx = f(x)
plt.plot(x, Mq, label='M*q(x)')
plt.plot(x, fx, label='f(x)')
plt.plot(x,q(x),label='q(x)')
plt.xlabel('x')
plt.ylabel('Value')
plt.title('M*q(x), f(x) and q(x) Curves')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Problem 2.2
# Rejection sampling
def rejection_sampling(f, q, q_sample, M, N):
    """
    f: target distribution
    g: proposal distribution
    g_sample: sample from g
    M: constant
    N: number of samples
    """
    samples = []
    for i in range(N):
        x = q_sample()
        u = np.random.uniform(0, M*q(x))
        if u <= f(x):
            samples.append(x)
    return samples
samples_from_rejection_sampling = rejection_sampling(f, q, q_sample, M, 1000000)
```

```python
# plot histogram
plt.hist(samples_from_rejection_sampling, bins=360, density=True, alpha=0.7, color='g',label='Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Fitted Normal Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def estimate_expectation_squared(samples):
    squared_samples = np.square(samples)
    expectation_squared = np.mean(squared_samples)
    return expectation_squared
samples = samples_from_rejection_sampling
estimated_expectation_squared = estimate_expectation_squared(samples)
print("Estimated E(X^2):", estimated_expectation_squared)
```
